# CIC IIoT Dataset 2025 - Data Exploration

Exploring the processed CSV files to understand dataset structure and characteristics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create figures directory
os.makedirs('figures', exist_ok=True)

sns.set_style('whitegrid')
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 100)
```

## 1. Load Data

```python
# Load attack data
attack_df = pd.read_csv('dataset/attack_samples_1sec.csv')
print(f"Attack data shape: {attack_df.shape}")
print(f"Memory usage: {attack_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
# Load benign data
benign_df = pd.read_csv('dataset/benign_samples_1sec.csv')
print(f"Benign data shape: {benign_df.shape}")
print(f"Memory usage: {benign_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

## 2. Basic Structure

```python
# Check column names
print("Attack columns:")
print(attack_df.columns.tolist())
print(f"\nTotal columns: {len(attack_df.columns)}")
```

```python
# Check if columns match
print(f"Benign columns match attack columns: {list(benign_df.columns) == list(attack_df.columns)}")
if list(benign_df.columns) != list(attack_df.columns):
    print("\nColumn differences:")
    print(f"Only in attack: {set(attack_df.columns) - set(benign_df.columns)}")
    print(f"Only in benign: {set(benign_df.columns) - set(attack_df.columns)}")
```

```python
# First few rows of attack data
attack_df.head()
```

```python
# First few rows of benign data
benign_df.head()
```

## 3. Data Types and Missing Values

```python
# Data types
print("Attack data types:")
print(attack_df.dtypes.value_counts())
print("\nBenign data types:")
print(benign_df.dtypes.value_counts())
```

```python
# Missing values in attack data
missing_attack = attack_df.isnull().sum()
missing_attack = missing_attack[missing_attack > 0].sort_values(ascending=False)
if len(missing_attack) > 0:
    print("Missing values in attack data:")
    print(missing_attack)
    print(f"\nPercentage: {(missing_attack / len(attack_df) * 100).round(2)}%")
else:
    print("No missing values in attack data")
```

```python
# Missing values in benign data
missing_benign = benign_df.isnull().sum()
missing_benign = missing_benign[missing_benign > 0].sort_values(ascending=False)
if len(missing_benign) > 0:
    print("Missing values in benign data:")
    print(missing_benign)
    print(f"\nPercentage: {(missing_benign / len(benign_df) * 100).round(2)}%")
else:
    print("No missing values in benign data")
```

## 4. Target/Label Distribution

```python
# Find label column (usually 'label', 'Label', 'attack_type', etc.)
label_candidates = [col for col in attack_df.columns if 'label' in col.lower() or 'attack' in col.lower() or 'type' in col.lower()]
print(f"Potential label columns: {label_candidates}")
```

```python
# Check unique values in potential label columns
for col in label_candidates:
    print(f"\n{col}:")
    print(f"Unique values: {attack_df[col].nunique()}")
    print(f"Sample values: {attack_df[col].unique()[:10]}")
```

## 5. Basic Statistics

```python
# Numeric columns statistics for attack data
numeric_cols = attack_df.select_dtypes(include=[np.number]).columns
print(f"Number of numeric features: {len(numeric_cols)}")
attack_df[numeric_cols].describe()
```

```python
# Numeric columns statistics for benign data
numeric_cols_benign = benign_df.select_dtypes(include=[np.number]).columns
print(f"Number of numeric features: {len(numeric_cols_benign)}")
benign_df[numeric_cols_benign].describe()
```

## 6. Check for Categorical/Object Columns

```python
# Non-numeric columns in attack data
non_numeric_attack = attack_df.select_dtypes(exclude=[np.number]).columns
print(f"Non-numeric columns in attack data ({len(non_numeric_attack)}):")
for col in non_numeric_attack:
    print(f"  {col}: {attack_df[col].nunique()} unique values")
    if attack_df[col].nunique() < 20:
        print(f"    Values: {attack_df[col].unique()}")
```

```python
# Non-numeric columns in benign data
non_numeric_benign = benign_df.select_dtypes(exclude=[np.number]).columns
print(f"Non-numeric columns in benign data ({len(non_numeric_benign)}):")
for col in non_numeric_benign:
    print(f"  {col}: {benign_df[col].nunique()} unique values")
    if benign_df[col].nunique() < 20:
        print(f"    Values: {benign_df[col].unique()}")
```

## 7. Attack Type Distribution

```python
# Attack categories (label2)
attack_categories = attack_df['label2'].value_counts()
print("Attack categories distribution:")
print(attack_categories)
print(f"\nPercentage:")
print((attack_categories / len(attack_df) * 100).round(2))
```

```python
# Visualize attack categories
plt.figure(figsize=(10, 6))
attack_categories.plot(kind='bar')
plt.title('Attack Category Distribution')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('figures/attack_category_distribution.pdf', dpi=300, bbox_inches='tight')
print("Saved: figures/attack_category_distribution.pdf")
plt.show()
```

```python
# Top 15 specific attack types (label3)
top_attacks = attack_df['label3'].value_counts().head(15)
print("Top 15 specific attack types:")
print(top_attacks)
```

```python
# Visualize top 15 attacks
plt.figure(figsize=(12, 6))
top_attacks.plot(kind='barh')
plt.title('Top 15 Specific Attack Types')
plt.xlabel('Count')
plt.ylabel('Attack Type')
plt.tight_layout()
plt.savefig('figures/top_15_attack_types.pdf', dpi=300, bbox_inches='tight')
print("Saved: figures/top_15_attack_types.pdf")
plt.show()
```

## 8. Class Balance

```python
# Overall class balance (benign vs attack)
total_benign = len(benign_df)
total_attack = len(attack_df)
total = total_benign + total_attack

print("Overall class distribution:")
print(f"Benign: {total_benign:,} ({total_benign/total*100:.2f}%)")
print(f"Attack: {total_attack:,} ({total_attack/total*100:.2f}%)")
print(f"\nClass ratio (Attack:Benign): 1:{total_benign/total_attack:.2f}")
```

```python
# Visualize class balance
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Pie chart
axes[0].pie([total_benign, total_attack], labels=['Benign', 'Attack'], 
            autopct='%1.1f%%', startangle=90)
axes[0].set_title('Binary Class Distribution')

# Bar chart
axes[1].bar(['Benign', 'Attack'], [total_benign, total_attack])
axes[1].set_title('Binary Class Distribution')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('figures/class_distribution.pdf', dpi=300, bbox_inches='tight')
print("Saved: figures/class_distribution.pdf")
plt.show()
```

## 9. Feature Characteristics

```python
# Check for constant features (same value for all samples)
constant_features = []
for col in numeric_cols:
    if attack_df[col].nunique() == 1 and benign_df[col].nunique() == 1:
        constant_features.append(col)

if len(constant_features) > 0:
    print(f"Constant features (same value in both datasets): {len(constant_features)}")
    print(constant_features[:10])
else:
    print("No constant features found")
```

```python
# Check for features with many zeros
zero_heavy_features = []
for col in numeric_cols:
    zero_pct = (attack_df[col] == 0).sum() / len(attack_df) * 100
    if zero_pct > 80:
        zero_heavy_features.append((col, zero_pct))

print(f"Features with >80% zeros in attack data: {len(zero_heavy_features)}")
if len(zero_heavy_features) > 0:
    print("\nTop 10:")
    for feat, pct in sorted(zero_heavy_features, key=lambda x: x[1], reverse=True)[:10]:
        print(f"  {feat}: {pct:.1f}%")
```
